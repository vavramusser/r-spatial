### 2.1 Inspect Data

What to look for when inspecting spatial data:

1. **Geometry Type** - Are these points, lines, or polygons?
2. **Number of Features** - How many spatial objects are in the dataset?
3. **Coordinate Reference System (CRS)** - What projection is being used?
4. **Attributes** - What variables are associated with each feature?
5. **Bounding Box** - What geographic extent does the data cover?

The USA States data contains 52 polygon features (50 states, Washington DC, and Puerto Rico) with four attributes including the [FIPS code](https://en.wikipedia.org/wiki/Federal_Information_Processing_Standard_state_code), 2-letter USPS code, name, and Census region for each state.  The data is projected in the USA Contiguous Albers Equal Area Conic CRS.

### 2.2 Modify Data for Simplified Mapping

Let's make a slightly simpler version of the data for the sake of this project.  We'll remove Alaska, Hawaii, and Puerto Rico to make our maps simpler and recode the Census region attribute to reflect the region names.

```r
# remove Alaska, Hawaii, and Puerto Rico
# and recode the region variable
usa_states <- usa_states %>%
    filter(!(STUSPS10 %in% c("AK", "HI", "PR"))) %>%
    mutate(REGION10 = recode(REGION10,
                             "1" = "Northeast",
                             "2" = "Midwest",
                             "3" = "South",
                             "4" = "West"))
```

## 3. Quick Exploratory Mapping with Base R

Before getting into advanced visualization techniques, itâ€™s helpful to know the simplest approach: base R's `plot()` function.  While `plot()` lacks the customization options of other methods, it can be used to create very quick visualizations.  This can be useful in the early stages of spatial analysis to get a sense of the structure, extent, and contents of a spatial dataset.

### Why Use Base R plot()?

The `plot()` function is ideal for:
- **Quick Data Checks** - Verify that data loaded correctly and is what's expected
- **Initial Exploration** - Get a sense of spatial patterns, geographic extent, etc.
- **Checking CRS** - Ensure projections look correct
- **Rapid Feedback** - Quick checks between data cleaning and engineering steps
- **Lightweight** - Works with base R, no additional packages

### When NOT to Use plot()

- Publication-quality figures
- Complex multi-layer maps
- Precise control over styling
- Integration with other ggplot2 visualizations

### 3.1 Plotting Geometry Only

The simplest map is one that shows only the spatial boundaries of your data with no data attributes. This is useful for confiming the data loaded correctly, checking the geographic extent, and ensuring that the data is properly projected and in the right CRS.

```r
# simple geometry plot
plot(st_geometry(usa_states))
```

### 3.2 Adding Basic Styling

Base R plots do have some customization options.

```r
# add custom colors, a title, and axes
plot(st_geometry(usa_states), 
     col = "lightgreen",                        # fill color
     border = "navy",                           # border color
     main = "United States State Boundaries",   # title
     axes = TRUE)                               # show axes
```

### 3.3 Plotting with Attributes

Maps are most powerful when they show data attributres and patterns.  You can use plot() to quickly visualize attributes by specifying a column name.

```r
# visualize the Census region (REGION10) attribute
plot(usa_states["REGION10"],                # plot the region attribute
     border = "navy",                       # border color
     main = "US States by Census Region",   # title
     key.pos = 1,                           # legend position (1 = below the plot)
     axes = TRUE)                           # show axes
```

### 3.4 Plotting Multiple Attributes

You can even plot multiple attributes at the same time to compare patterns.

```r
# plot both the Census Region (REGION10) and State USPS (STUSPS10) code attributes in a grid
plot(usa_states[c("REGION10", "STUSPS10")])
```

### Key Takeaways: Base R plot()

**Pros:**
- Extremely fast to write and execute
- No additional packages required (works with sf alone)
- Good for quick data checks
- Can plot multiple attributes quickly

**Cons:**
- Limited customization options
- Not suitable for publication-quality figures
- Harder to create complex multi-layer maps
- Less control over legends and styling

**Best use cases:**
- Initial data exploration

## 4. Customized Mapping with ggplot2

While base R *plot()* is great for quick exploration, [*ggplot2*](https://cran.r-project.org/web/packages/ggplot2/index.html) offers the customization and aesthetic control needed for publication-quality maps.

### Why ggplot2 for Mapping?

**Advantages:**
- **Consistent Syntax** - If you use ggplot2 to create data visualuzations, you already know how to make maps!
- **Layering** - Easily combine multiple spatial and non-spatial elements
- **Customization** - Control over every visual element
- **Publication-Ready** - Produces high-quality maps for publication
- **Reproducibility** - Code-based approach ensures consistent results

**When to use ggplot2:**
- Creating maps for publications or presentations
- Combining maps with other ggplot2 visualizations
- Need precise control over styling and layout
- Creating multi-panel (faceted) maps

### 4.1 Basic ggplot2 Map

A great first step in using [*ggplot2*](https://cran.r-project.org/web/packages/ggplot2/index.html) for spatial visualization is to plot only the geometries in your spatial object without any attributes or variables. This allows you to quickly assess the shapes, locations, and layout of your data.

The foundation of any ggplot2 map is the combination of [*ggplot*](https://rdrr.io/cran/ggplot2/man/ggplot.html) and [*geom_sf*](https://rdrr.io/cran/ggplot2/man/ggsf.html) which is used to render spatial objects stored in the [simple features (sf)](https://cran.r-project.org/web/packages/sf/index.html) structure.  This creates a simple map showing just the geometries.  When no fill or color aesthetic is specified, [*ggplot2*](https://cran.r-project.org/web/packages/ggplot2/index.html) will use default values to outline polygons or plot points.

```r
# basic ggplot2 map using geom_sf
ggplot(data = usa_states) +   # use the usa_states data
    geom_sf() +               # add spatial geometries
    theme_minimal()           # minimal theme
```

**Understanding the code:**

- `ggplot(data = usa_states)` - Initialize plot with our spatial data
- `geom_sf()` - Add spatial geometries (automatically detects geometry column)
- `theme_minimal()` - Apply a clean, minimal theme

This three-line command creates a basic map. From here, we can add layers of complexity.

### 4.2 Mapping Data to Visual Properties

The power of ggplot2 comes from mapping data attributes to visual aesthetics like color, fill, size, or transparency. This is done using the [*aes*](https://rdrr.io/cran/ggplot2/man/aes.html) function.

Once you've mapped the raw geometries, the next step is to visualize the attributes associated with those spatial features. This is the essence of thematic mapping - using color, size, or shape to represent data values tied to geography.

Using [*ggplot2*](https://cran.r-project.org/web/packages/ggplot2/index.html), we can map variables by assigning them to aesthetics like fill, color, or size. For example, we might shade polygons based on population density, or color points based on category or count.

In this section, youâ€™ll learn how to:

- Use the aes() function within geom_sf() to link spatial data attributes to visual elements.
- Control how data-driven attributes are mapped to the appearance of the plot.
- Create visual representations of social or environmental phenomena using real data.

These maps are often called choropleths (for polygons) or graduated symbol maps (for points)â€”and they serve as powerful tools for pattern detection and storytelling.

#### Mapping Categorical Data

```r
# visualize the categorical Census region (REGION10) attribute
ggplot(data = usa_states) +           # use the usa_states data
    geom_sf(aes(fill = REGION10)) +   # fill polygons using the Census region (REGION10) attribute
    theme_minimal()                   # minimal theme
```

#### Mapping Continuous Data

For continuous variables (like population or area), ggplot2 automatically creates a color gradient.

```r
# first, lets calculate each state's area as a new column (attribute)
# the original data is in square meters, so let's also convert the final version to square miles
usa_states$area_sq_mi <- as.numeric(st_area(usa_states)) / 2589988   # convert to square miles

# visualize the continuous area_sq_mi attribure
ggplot(data = usa_states) +
  geom_sf(aes(fill = area_sq_mi)) +                # fill polygons using the area in square miles (area_sq_mi) attribute
  scale_fill_viridis_c(name = "Area (miÂ²)",        # use the continuous viridis color palette and add a legend label
                       labels = scales::comma) +   # include commas in numbers in the legend
  theme_minimal()
```

### 4.3 Adding Labels and Annotations

Clear labels are essential for effective communication. The [*labs*](https://rdrr.io/cran/ggplot2/man/labs.html) function controls titles, subtitles, captions, and legend labels:

```r
# map with comprehensive labels
ggplot(data = usa_states) +
  geom_sf(aes(fill = REGION10)) +
  labs(                                                                 # specify labels and titles
    title = "United States Census Regions",                             # add an overall title
    subtitle = "Based on U.S. Census Bureau regional classification",   # add a subtitle
    caption = "Data source: IPUMS NHGIS",                               # add a caption with data source information
    fill = "Census Region"                                              # add a legend label
  ) +
  theme_minimal()
```

### 5.4 Customizing Borders and Fills

You have complete control over both fill colors and border properties:

Effective use of color can significantly improve the readability and impact of your maps. In this section, we explore how to fine-tune the color palettes used for attribute-based mapping.

Whether you're mapping continuous variables (like income or temperature) or categorical ones (like land use types), choosing the right palette is essential for conveying your message clearly and accessibly.

Here you'll learn how to:

- Apply color palettes using scale_fill_*() and scale_color_*() functions.
- Choose appropriate palettes for sequential, diverging, and qualitative data.
- Use colorblind-friendly and perceptually uniform palettes, such as those from the viridis and RColorBrewer packages.

```r
# customized borders and fill colors
ggplot(data = usa_states) +
  geom_sf(aes(fill = REGION10),
          color = "white",              # border color
          size = 0.3,                   # border thickness (1 = default)
          alpha = 0.8) +                # transparency (1 = fully opaque; 0 = fully transparent)
  labs(title = "U.S. Census Regions",
       fill = "Region") +
  theme_void() +
  theme(legend.position = "bottom")     # legend position
```


Good color choices are not just aesthetic, they also help ensure your maps are readable and interpretable by a broader audience, including those with color vision deficiencies.

Choosing appropriate color palettes is crucial when visualizing spatial data. Effective palettes help communicate the underlying data clearly and accurately, especially when color is used to represent continuous or categorical attributes.

**Best practices:**
- Use sequential palettes (e.g., Blues, Viridis) for continuous data.
- Use qualitative palettes (e.g., Set1, Dark2) for categorical data.
- Choose colorblind-friendly palettes where possible.
- Use perceptually uniform palettes (like `viridis`) for improved readability.

### 4.5 Controlling Map Projection and Extent

The *coord_sf* function gives you control over the map's projection and spatial extent.  Here let's focus on the state of Minnesota by including the latitude (Y) and longitude (X) limits of the bounding box for the state.

```r
# focus on only the contiguous 48 states with Albers projection
ggplot(data = usa_states) +
  geom_sf() +
  coord_sf(xlim = c(-97.24, -89.48),   # longitude limits for Minnesota
           ylim = c(43.50, 49.38),     # latitude limits for Minnesota
           crs = 4326) +               # project the data to WGS84 on the fly (degree coordinates)
  labs(title = "Minnesota") +
  theme_minimal()
```

### 4.6 Creating Small Multiple Maps (Faceting)

One of ggplot2's powerful features is faceting - creating multiple maps in a grid:

```r
# create separate maps for each region with fixed scales
ggplot(data = usa_states) +
  geom_sf(aes(fill = area_sq_mi), color = "white", size = 0.2) +
  facet_wrap(~REGION10, ncol = 2) +
  scale_fill_viridis_c(name = "Area (sq mi)",
                       labels = scales::comma) +
  labs(title = "State Area by U.S. Census Region",
       subtitle = "Each panel shows states within a census region") +
  theme_void() +   # Clean look without axes
  theme(legend.position = "bottom",
        legend.key.width = unit(2, "cm"),
        strip.text = element_text(size = 11, face = "bold"),
        plot.title = element_text(size = 14, hjust = 0.5, face = "bold"),
        plot.subtitle = element_text(size = 10, hjust = 0.5, margin = margin(b = 10))
  )
```

### 4.7 Color Palettes in ggplot2

Choosing appropriate colors is crucial for effective maps. Let's explore different palette options:

#### Viridis Palettes (Recommended for Most Uses)

For reference, you can [view all of the viridis color palettes here](https://search.r-project.org/CRAN/refmans/viridisLite/help/figures/viridis-scales.png).

```r
# Viridis continuous scale
ggplot(data = usa_states) +
  geom_sf(aes(fill = area_sq_mi)) +
  scale_fill_viridis_c(option = "plasma",          # REVERSED plasma continuous palette yellow (low) to purple (high)
                       direction = -1,             # specfiy direction (-1 = reversed palette)
                       name = "Area (sq mi)") +
  labs(title = "State Area - Viridis Plasma Palette") +
  theme_minimal()
```

#### ColorBrewer Palettes

For reference, you can [view all of the ColorBrewer palettes here](https://r-graph-gallery.com/38-rcolorbrewers-palettes_files/figure-html/thecode-1.png).

```r
# ColorBrewer discrete palette
ggplot(data = usa_states) +
  geom_sf(aes(fill = REGION10)) +
  scale_fill_brewer(palette = "Pastel1",
                    name = "Region") +
  labs(title = "Census Regions - ColorBrewer Pastel Set 1") +
  theme_minimal()
```

```r
# ColorBrewer continuous palette
ggplot(data = usa_states) +
  geom_sf(aes(fill = area_sq_mi)) +
  scale_fill_distiller(palette = "YlGn",          # yellow-green continuous palette yellow (low) to green (high)
                       direction = 1,             # specfiy direction (-1 = reversed palette)
                       name = "Area (sq mi)") +
  labs(title = "State Area - ColorBrewer Yellow-Green Continuous Palette") +
  theme_minimal()
```

```r
# ColorBrewer diverging palette
ggplot(data = usa_states) +
  geom_sf(aes(fill = area_sq_mi)) +
  scale_fill_distiller(palette = "Spectral",      # REVERSED spectral diverging palette blue (low) to red (high)
                       direction = -1,            # specfiy direction (-1 = reversed palette)
                       name = "Area (sq mi)") +
  labs(title = "State Area - ColorBrewer Spectral Diverging Palette") +
  theme_minimal()
```

#### Manual Color Assignment

For help getting hex codes for individual colors, try a [hex color picker](https://htmlcolorcodes.com/color-picker).

```r
# Manually specified colors
region_colors <- c(
  "Northeast" = "#E41A1C",
  "Midwest" = "#377EB8",
  "South" = "#4DAF4A",
  "West" = "#984EA3"
)

ggplot(data = usa_states) +
  geom_sf(aes(fill = REGION10)) +
  scale_fill_manual(values = region_colors,
                    name = "Region") +
  labs(title = "Census Regions - Custom Colors") +
  theme_minimal()
```

## 5. Thematic Mapping with tmap

The `tmap` package is specifically designed for creating thematic maps. It offers an intuitive syntax that's map-focused and can produce both static and interactive maps with minimal code changes.

### Why tmap?

**Advantages:**
- **Map-specific design** - Built from the ground up for cartography
- **Intuitive syntax** - Layer-based like ggplot2 but optimized for maps
- **Quick mode switching** - Switch between static and interactive with one line
- **Smart defaults** - Cartographic best practices built in
- **Comprehensive features** - Scale bars, north arrows, insets all built-in
- **Flexible layouts** - Easy to create map layouts and arrangements

**When to use tmap:**
- Creating thematic maps is your primary goal
- Need both static and interactive versions
- Want cartographic elements (scale bars, north arrows)
- Prefer concise, map-focused syntax
- Creating map layouts with multiple maps

### 5.1 Basic tmap Map

The basic structure of a tmap map is straightforward:

```r
# Basic tmap
tm_shape(usa_states) +
  tm_borders()
```

**Understanding tmap syntax:**

- `tm_shape()` - Define the spatial object to map
- `tm_borders()` - Draw just the boundaries
- Additional `tm_*` functions add layers and styling

This two-line approach creates a clean, simple map.

### 5.2 Mapping Attributes with tmap

```r
# Map categorical data
tm_shape(usa_states) +
  tm_polygons("REGION10",
              title = "Census Region",
              palette = "Set2")
```

### 5.3 Adding Map Elements

tmap makes it easy to add professional cartographic elements:

```r
# Map with scale bar, north arrow, and title
tm_shape(usa_states) +
  tm_polygons("REGION10",
              title = "Region",
              palette = "Set3") +
  tm_scale_bar(position = c("left", "bottom")) +
  tm_compass(position = c("right", "top")) +
  tm_layout(title = "United States Census Regions",
            title.position = c("left", "top"),
            legend.position = c("right", "bottom"),
            frame = TRUE)
```

### 5.4 Using Built-in Map Styles

tmap comes with several pre-built professional styles:

```r
# Classic style
tm_shape(usa_states) +
  tm_polygons("REGION10") +
  tm_style("classic")
```

```r
# Natural style
tm_shape(usa_states) +
  tm_polygons("REGION10") +
  tm_style("natural")
```

```r
# Cobalt style (dark theme)
tm_shape(usa_states) +
  tm_polygons("REGION10") +
  tm_style("cobalt")
```

### 5.5 Switching to Interactive Mode

One of tmap's most powerful features is the ability to switch between static and interactive modes:

```r
# Switch to interactive mode
tmap_mode("view")

# Same code now produces an interactive map!
tm_shape(usa_states) +
  tm_polygons("REGION10",
              title = "Census Region",
              palette = "Set2",
              popup.vars = c("State" = "STATENAM",
                            "Region" = "REGION10"))
```

```r
# Switch back to static mode
tmap_mode("plot")
```

### Key Takeaways: tmap

**Pros:**
- Specifically designed for thematic mapping
- Intuitive, map-focused syntax
- Easy switching between static and interactive
- Built-in cartographic elements (scale bars, north arrows)
- Professional default styling
- Excellent for creating map layouts

**Cons:**
- Less flexible than ggplot2 for non-map visualizations
- Smaller ecosystem than ggplot2
- Less integration with non-spatial tidyverse workflows

**Best use cases:**
- Thematic mapping is your primary focus
- Need both static and interactive versions
- Want cartographic best practices by default
- Creating map layouts and arrangements
- Rapid prototyping of map designs

## 6.  Comparing Mapping Approaches

There are multiple ways to create maps in R, each with its own strengths depending on your goals, workflow, and audience.  We've now explored three different ways to create maps in R. Let's compare them side by side:

- `plot()` is fast and simple - great for quick data checks and exploratory work.
- `ggplot2` is highly customizable and integrates well with tidyverse data pipelines.
- `tmap` provides easy thematic mapping with high-level functions and supports both static and interactive mapping modes.

### Feature Comparison

| Feature | base R plot() | ggplot2 | tmap |
|---------|--------------|---------|------|
| **Learning Curve** | Easy | Moderate | Moderate |
| **Code Verbosity** | Minimal | Moderate | Minimal |
| **Customization** | Limited | Extensive | Moderate-Extensive |
| **Map Elements** | Basic | Via extensions | Built-in |
| **Interactivity** | No | Via plotly | Built-in |
| **Faceting** | Manual | Excellent | Good |
| **Publication Quality** | Basic | Excellent | Excellent |
| **Speed** | Very Fast | Moderate | Moderate |
| **Tidyverse Integration** | No | Yes | Partial |

### When to Use Each Approach

#### Use base R `plot()` when:
- Doing quick exploratory analysis
- Checking if data loaded correctly
- Working interactively in the console
- Speed is more important than aesthetics
- You need a quick visual check, not a final product

#### Use ggplot2 when:
- Creating publication-quality figures
- Integrating with other ggplot2 visualizations
- Need extensive customization options
- Working within tidyverse workflows
- Creating faceted (small multiple) maps
- Want consistency across all your visualizations

#### Use tmap when:
- Thematic mapping is your primary goal
- Need both static and interactive versions
- Want cartographic elements (scale bars, north arrows) easily
- Prefer concise, map-specific syntax
- Creating map layouts and arrangements
- Rapid prototyping of map designs

### Practical Workflow Recommendation

In practice, many R users combine all three approaches:

1. **Start with `plot()`** for initial data exploration
2. **Use `tmap`** for creating and iterating on thematic maps
3. **Switch to `ggplot2`** when you need to integrate maps with other visualizations or need very specific customization

There's no single "best" approachâ€”the right tool depends on your specific needs and context.

## Quick Reference Code

Here's a clean version of the essential code from this notebook for easy copy-paste:

```r
# === SETUP ===
library(sf)
library(ggplot2)
library(tmap)
library(viridis)

# Load data
usa_states <- st_read("ipums_nhgis_states.shp")

# === BASE R PLOT ===
plot(st_geometry(usa_states))   # boundary only
plot(usa_states["REGION10"])    # map the region attribute attribute

# === GGPLOT2 ===
# Basic map
ggplot(usa_states) +
  geom_sf() +                   # geometries only
  theme_minimal()               # mininmal theme

# Categorical data
ggplot(usa_states) +
  geom_sf(aes(fill = REGION10)) +    # map the region attribute (categorical)
  scale_fill_viridis_d() +           # viridis color palette
  labs(title = "Census Regions") +   # add a title
  theme_minimal()                    # minimal theme

# Continuous data
ggplot(usa_states) +
  geom_sf(aes(fill = AREA)) +                 # map the area attribute (continuous)
  scale_fill_viridis_c(option = "plasma") +   # viridis color palette - plasma
  labs(title = "State Area") +                # add a title
  theme_minimal()                             # minimal theme

# === TMAP ===
# Static mode
tmap_mode("plot")

# Basic map
tm_shape(usa_states) +            
  tm_polygons("REGION10",               # geometries only
              title = "Region",         # add a label 
              palette = "Set2") +       # add a color palette
  tm_layout(title = "Census Regions")   # add a title

# Interactive mode
tmap_mode("view")
tm_shape(usa_states) +
  tm_polygons("REGION10")               # plot the region attribute in interactive mode 
```
